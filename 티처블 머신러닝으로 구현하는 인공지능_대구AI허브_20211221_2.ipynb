```python
# 예외처리 예제 3 - 2

no1 = int(input('숫자를 입력하세요: '))
no2 = int(input('숫자를 입력하세요: '))
try :
  print(no1 / no2)
except ZeroDivisionError :
  print('0으로 나눌 수 없습니다.')
# ZeroDivisionError가 발생
```

```python
# 예외처리 예제 3 - 3(어떤 오류가 발생할 지 모르는 경우)

no1 = int(input('숫자를 입력하세요: '))
no2 = int(input('숫자를 입력하세요: '))
try :
  print(no1 / no2)
except  :
  print('값을 확인해주세요.')
```

```python
# 예외처리 예제 4

try:
  no = int(input('숫자만 입력하세요: '))
except :
  print('꼭 숫자만 입력해주세요.')
else :
  print('입력하신 숫자는 {}입니다.' .format(no))
```

```python
# 예외처리 - finally 사용

try :
  num = int(input('숫자를 입력하세요.'))
except ValueError :
  print('숫자가 아닙니다.')
else :
  print('입력한 숫자는 %s입니다.' %num)
finally :
  print('finally는 무조건 실행됩니다.')
```

```python
# merge 실습용 데이터 생성 1

import pandas as pd 
df_1 = {'학번' : ['1001','1002','1003'],
        '이름' : ['홍길동','일지매','전우치'],
        '키' : [180,175,168]}
stu_1 = pd.DataFrame(df_1)
stu_1
```

```python
# merge 실습용 데이터 생성 2

import pandas as pd 
df_2 = {'학번' : ['1001','1002','1003'],
        '이름' : ['홍길동','일지매','전우치'],
        '몸무게' : [80,75,68]}
stu_2 = pd.DataFrame(df_2)
stu_2
```

```python
# 기본값 - merge 할 기준을 지정

pd.merge(stu_1, stu_2, how = 'left', left_on = '이름', right_on = '이름')
```

```python
# 기본값 - merge 할 기준을 지정

pd.merge(stu_1, stu_2, how = 'outer', left_on = '이름', right_on = '이름')
```

```python
pd.concat([stu_1 , stu_2],  axis = 0)
```

```python
pd.concat([stu_1 , stu_2],  axis = 1)
```
